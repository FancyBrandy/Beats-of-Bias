```python
# importing required libraries
import pandas as pd
import numpy as np
import re
```

```python
path = '/content/drive/MyDrive/Praktikum - NLP Applications/Datasets/genius_and_wasabi/concatenated_chunks.csv'

data = pd.read_csv(path, low_memory=False)
```

```python
# Removing explicit song structure as it is not important information

def remove_explicit_song_structure():
  pattern = r'\[.+?\]'
  updated_lyrics = []

  for lyrics in data['lyrics']:
    cleaned_lyrics = re.sub(pattern, '', lyrics)
    updated_lyrics.append(cleaned_lyrics)

  data['updated_lyrics'] = updated_lyrics

remove_explicit_song_structure()
data.head()
```
