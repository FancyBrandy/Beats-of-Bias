```python
# importing required libraries
import pandas as pd
import numpy as np
# import dask.dataframe as dd
```

```python
path = '/content/drive/MyDrive/Praktikum - NLP Applications/Datasets/genius_and_wasabi/'

def create_chunks_and_store():

  # Load wasabi_artists.csv
  artists_wasabi = pd.read_csv('/content/drive/MyDrive/Praktikum - NLP Applications/Datasets/wasabi_artists.csv', low_memory= False)
  artists_columns = ['gender', 'name']
  artists_wasabi = artists_wasabi[artists_columns]

  # Load song_lyrics.csv
  chunksize = 1000000
  genius_songs_pandas = pd.read_csv('/content/drive/MyDrive/Praktikum - NLP Applications/Datasets/genius_song_lyrics.csv', low_memory=False, chunksize = chunksize)

  for i, chunk in enumerate(genius_songs_pandas):
    # Filter non-english songs
    genius_songs = chunk[chunk['language'] == 'en']

    # Inner join with wasabi_artists
    songs_and_artists = pd.merge(genius_songs, artists_wasabi, how = 'inner', left_on = 'artist', right_on = 'name')

    # Filter out gender
    songs_and_artists = songs_and_artists.dropna(subset = ['gender'])
    songs_and_artists = songs_and_artists[songs_and_artists['gender'] != 'Other']

    # Drop unnecessary columns
    songs_and_artists = songs_and_artists.drop(columns=['language_cld3', 'language_ft', 'id', 'name'])
    songs_and_artists = songs_and_artists.rename(columns = {'tag': 'genre'})

    # Store chunk as csv
    data_path = path + 'chunk_' + str(i+1) + '.csv'
    songs_and_artists.to_csv(data_path)
```

```python
create_chunks_and_store()
```

```python
chunk1 = pd.read_csv(path + 'chunk_1.csv')
chunk2 = pd.read_csv(path + 'chunk_2.csv')
chunk3 = pd.read_csv(path + 'chunk_3.csv')
chunk4 = pd.read_csv(path + 'chunk_4.csv')
chunk5 = pd.read_csv(path + 'chunk_5.csv')
chunk6 = pd.read_csv(path + 'chunk_6.csv')

chunks = [chunk1, chunk2, chunk3, chunk4, chunk5, chunk6]
df = pd.concat(chunks)
df.to_csv(path + 'concatenated_chunks.csv')
```
